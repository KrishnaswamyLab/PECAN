```python
datasets = (spiral, blobs, moons, nested_circles, barbell, double_annulus, hyperuniform_circle, hyperuniform_ellipse, linked_hyperuniform_circles, petals, poisson_process, euclidean_disk)
for dataset in datasets:
    # Create a 4x5 grid of subplots
    fig, axes = plt.subplots(4, 5, figsize=(20, 16))

    epsilons = np.arange(1, 21) * 0.051
    alpha = 10
    # Assuming petals is a function that you've defined to generate data
    for i, epsilon in enumerate(epsilons):
        row = i // 5  # Integer division to get row index
        col = i % 5   # Remainder to get column index
        X, _ = dataset(128, random_state=42)
        P, d = make_diffusion_d(X, alpha=2, epsilon=epsilon)
        visuallize_diffusion_map_subplot_correct(X, P, d, ax=axes[row, col], name=dataset.__name__, epsilon=epsilon, alpha=alpha, t=20)
    plt.suptitle(dataset.__name__)
    plt.tight_layout()
    plt.show()
```
