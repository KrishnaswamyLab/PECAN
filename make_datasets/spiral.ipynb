```python
import numpy as np
import matplotlib.pyplot as plt

def generate_spiral(N, A=1, omega=1, phi=0, t_min=0, t_max=10):
    """
    Generate a dataset of a spiral.

    Parameters:
    - N: Number of points
    - A: Amplitude
    - omega: Angular frequency
    - phi: Phase
    - t_min: Minimum time value
    - t_max: Maximum time value

    Returns:
    - X: 2D array of shape (N, 2), where each row is a point [x(t), y(t)]
    - t: 1D array of time values
    """
    # Generate time values
    t = np.linspace(t_min, t_max, N)
    
    # Generate the spiral points using the parametric equations
    x = A * t * np.cos(omega * t + phi)
    y = A * t * np.sin(omega * t + phi)
    
    # Combine x and y coordinates into a single 2D array
    X = np.column_stack((x, y))
    
    return X, t
```

```python
# Generate a test dataset with 500 points
X, t = generate_spiral(N=128, A=0.1, omega=2, phi=0, t_min=0, t_max=10)
plt.scatter(X[:,0],X[:,1])
np.savetxt("spiral.txt", X)
```

```python
import numpy as np

def generate_spiral2(N, A=1, omega=1, phi=0, alpha=0, t_min=0, t_max=10):
    """
    Generate a dataset of a spiral.

    Parameters:
    - N: Number of points
    - A: Initial amplitude
    - omega: Angular frequency
    - phi: Phase
    - alpha: Amplitude adjustment factor for reducing cycle distance
    - t_min: Minimum time value
    - t_max: Maximum time value

    Returns:
    - X: 2D array of shape (N, 2), where each row is a point [x(t), y(t)]
    - t: 1D array of time values
    """
    # Generate time values
    t = np.linspace(t_min, t_max, N)
    
    # Generate the spiral points using the modified parametric equations
    x = (A / (1 + alpha * t)) * t * np.cos(omega * t + phi)
    y = (A / (1 + alpha * t)) * t * np.sin(omega * t + phi)
    
    # Combine x and y coordinates into a single 2D array
    X = np.column_stack((x, y))
    
    return X, t
```

```python
# Generate a test dataset with 500 points
X, t = generate_spiral2(N=128, A=1, omega=4, phi=0, alpha=1, t_min=0, t_max=3)
plt.scatter(X[:,0],X[:,1])
np.savetxt("spiral2.txt", X)
```
